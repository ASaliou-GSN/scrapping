## Fonction génération identifiants

```python
import requests
from bs4 import BeautifulSoup
import json

def extraire_identifiant_coureur(chaine):
    debut_index = chaine.find("idp=")
    if debut_index != -1:
        debut_index += len("idp=")
        fin_index = chaine.find("&", debut_index)
        if fin_index != -1:
            identifiant = chaine[debut_index:fin_index]
            return identifiant
    return None

def remplir_json(num_page, fichier_json):
    #url = f"https://results.baa.org/2022/?page={num_page}&num_results=1000&pid=list"
    #url = f"https://results.chicagomarathon.com/2023/?page={num_page}&num_results=1000&pid=list"
    #url = f"https://chicago-history.r.mikatiming.com/2022/?pid=list&pidp=start&event_main_group=2017"
    url = f"https://chicago-history.r.mikatiming.com/2022/?pid=list&pidp=start"

    try:
        #response = requests.get(url)
        response = requests.post(url, data={'event_main_group': '2016'})
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")
            h4_tags = soup.find_all("h4")
            identifiants = []
            for h4_tag in h4_tags:
                href = h4_tag.find("a")["href"]
                identifiant = extraire_identifiant_coureur(href)
                if identifiant:
                    identifiants.append(identifiant)

            # Ouvrir le fichier JSON en mode lecture/écriture
            with open(fichier_json, "r+") as file:
                # Charger les identifiants déjà présents dans le fichier
                try:
                    ancien_contenu = json.load(file)
                except json.decoder.JSONDecodeError:
                    # Si le fichier est vide ou corrompu, initialiser avec une liste vide
                    ancien_contenu = []
                # Fusionner les nouveaux identifiants avec les anciens
                identifiants = list(set(ancien_contenu + identifiants))
                # Revenir au début du fichier pour écrire les identifiants fusionnés
                file.seek(0)
                # Écrire les identifiants fusionnés dans le fichier JSON
                json.dump(identifiants, file, indent=4)
                # Tronquer le fichier à la longueur actuelle pour effacer les données restantes
                file.truncate()
            print(f"Identifiants de la page {num_page} ajoutés avec succès dans le fichier {fichier_json}")
        else:
            print(f"La requête a échoué pour la page {num_page} avec le code : {response.status_code}")
    except Exception as e:
        print(f"Une erreur s'est produite : {str(e)}")

#remplir_json(27,"identifiants_add.json")
```

```python
def automatisation_identifiants(nombre_de_pages):
    for i in range(nombre_de_pages):
        remplir_json(i,"identifiants_Chicago_2017.json")


automatisation_identifiants(53)
```

## Récupérer le Json d'un seul coureur 

```python
import requests
from bs4 import BeautifulSoup
import json
import os



def get_final_json(runner_id,year):

    #with open(fichier_json, 'r') as f:
    #    data = json.load(f)


    #url = f"https://results.baa.org/{year}/?content=detail&fpid=list&pid=list&idp={runner_id}&lang=EN_CAP&event=R&num_results=1000&pidp=start&search_event="
    #url = "https://results.bostonmarathon.org/2023/?content=detail&fpid=list&pid=start&idp=9TGHS6FF181D5A&event=R"
    url = f"https://results.chicagomarathon.com/{year}/?content=detail&fpid=list&pid=list&idp={runner_id}&lang=EN_CAP&event=R&num_results=1000&pidp=start&search_event="

    response = requests.get(url)
    #print("connection réussie")
    soup = BeautifulSoup(response.content, "html.parser")

    # Extracting participant details
    participant_details = {}
    details_table = soup.find("div", class_="detail-box box-general").find("table")
    rows = details_table.find_all("tr")
    for row in rows:
        cells = row.find_all(["th", "td"])
        if len(cells) == 2:
            key = cells[0].text.strip()
            value = cells[1].text.strip()
            if key == "Name":
                # Splitting full name into parts
                parts = value.split(", ")
                if len(parts) == 2:
                    name, surname = parts
                elif len(parts) == 3:
                    name, _, surname = parts  # Utiliser la première et la troisième partie
                else:
                    # Gérer les cas où il y a moins de deux ou plus de trois éléments, si nécessaire
                    name = surname = "Unknown"  # Ou autre gestion d'erreur/logique appropriée

                participant_details["Name"] = name
                participant_details["Surname"] = surname
            else:
                participant_details[key] = value
                participant_details["identifiant"] = runner_id


    # Extracting totals
    totals = {}
    totals_table = soup.find("div", class_="detail-box box-totals").find("table")
    rows = totals_table.find_all("tr")
    for row in rows:
        cells = row.find_all(["th", "td"])
        if len(cells) == 2:
            key = cells[0].text.strip()
            value = cells[1].text.strip()
            totals[key] = value

    # Extracting race status
    race_status = {}
    status_table = soup.find("div", class_="detail-box box-state").find("table")
    rows = status_table.find_all("tr")
    for row in rows:
        cells = row.find_all(["th","td"])
        if len(cells) == 2:
            key = cells[0].text.strip()
            value = cells[1].text.strip()
            race_status[key] = value

    # Extracting splits
    splits = []
    splits_table = soup.find("div", class_="detail-box box-splits").find("table")
    rows = splits_table.find_all("tr")
    for row in rows[1:]:  # skipping the header row
        cells = row.find_all("td")
        desc = row.find("th", class_="desc").text.strip()  # Extract the name of the split
    
        split = {
            "Split": desc,  # Assign the name of the split
            "Time Of Day": cells[0].text.strip(),
            "Time": cells[1].text.strip(),
            "Diff": cells[2].text.strip(),
            #"Min/mile": cells[3].text.strip(),
            #"Miles/h": cells[4].text.strip()

        }
        splits.append(split)

    # Constructing JSON object
    participant_data = {
        "Participant Details": participant_details,
        "Totals": totals,
        "Race Status": race_status,
        "Splits": splits
    }

    return participant_data

    print(participant_data)

runner_id = "9TGG963828AFFB"   #ça peut servir si on oublie des gens (ça se voit quand on ajoute les sexe) pour 2023 : 2 personnes => 1 oubli et un gars supprimé des résultats 
get_final_json(runner_id,2023)
```

## Automatiser sur toutes les pages 


```python
def generate_full_json(ids_file, output_file, year):
    if os.path.exists(output_file):
        with open(output_file, 'r') as p:
            runners = json.load(p)
    else:
        runners = {}

    print("il y a bien ", len(runners), " coureurs dans mon fichier")

    with open(ids_file, 'r') as f:
        ids = json.load(f)
        start_index = len(runners)
        ids = ids[start_index:]  # Continue from where we left off
        print("il reste ",len(ids)," coureurs à faire")

    batch_size = 500
    batch_data = {}

    for i, runner_id in enumerate(ids):
        participant_data = get_final_json(runner_id, year)
        ranking = participant_data["Totals"]["Place Overall"]      #à modifier en fonction des marathons
        batch_data[ranking] = participant_data

        # Log progress
        print("le runner ", start_index + i + 1, "a été fait : ",runner_id)

        # Save in batches
        if (i + 1) % batch_size == 0 or i == len(ids) - 1:
            runners.update(batch_data)
            with open(output_file, 'w') as f:
                #test_1 = json.load(f)
                #print("la longueur de mon batch avant opération est :", len(test_1))
                json.dump(runners, f, indent=4)
                print("le dump a bien été fait")
                #test_2 = json.load(f)
                #print("la longueur de mon batch après opération est :", len(test_2))
            batch_data = {}  # Reset batch data after saving
            #start_index += i + 1  # Mise à jour de l'index de départ après chaque sauvegarde


# Example d'utilisation :
generate_full_json("identifiants_Chicago_2023.json", "Chicago_2023_V2.json", 2023)
```

## Update sex

```python
import json

def update_runner_gender(json_file):
    # Charger les données du fichier JSON
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Initialiser les compteurs
    count_non_binary = 0
    count_male = 0
    count_female = 0
    count_runner = 1

    # Parcourir chaque coureur par leur classement général
    for ranking in sorted(data, key=int):  # Assure que les clés sont triées numériquement
        runner = data[ranking]
        #place_ac = runner["Totals"]["Place (AC)"]
        place_ac = runner["Totals"]["Place (Category)"]
        #print(place_ac)
        #place_mw = int(runner["Totals"]["Place (M/W)"])
        place_mw = int(runner["Totals"]["Place (Gender)"])
        print(count_runner)

        if place_ac == "\u2013" or :
            runner["Participant Details"]["Sex"] = "NB"
            count_non_binary += 1
        elif count_male +1 == place_mw:          #hyper sexiste car si le compteur de femmes dépasse celui des hommes c'est faux
            runner["Participant Details"]["Sex"] = "M"
            count_male += 1
        elif count_female +1 == place_mw :
            runner["Participant Details"]["Sex"] = "W"
            count_female += 1
        else :
            print("error")
            runner["Participant Details"]["Sex"] = "W"
        count_runner+=1

    # Sauvegarder les modifications dans le fichier JSON
    with open(json_file, 'w') as file:
        json.dump(data, file, indent=4)

    return count_non_binary, count_male, count_female

# Exemple d'utilisation
nb, m, w = update_runner_gender("sorted_London_2024.json")
print(f"Non-binary: {nb}, Male: {m}, Female: {w}")
```

# Enlever les infos en trop

```python
import json

def remove_fields_from_splits(file_path):
    # Lire le fichier JSON
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Supprimer les champs "Min/mile" et "Miles/h" des splits
    for runner_id, runner_data in data.items():
        if "Splits" in runner_data:
            for split in runner_data["Splits"]:
                if "Min/mile" in split:
                    del split["Min/mile"]
                if "Miles/h" in split:
                    del split["Miles/h"]
                if "Time Of Day" in split:
                    del split["Time Of Day"]
    
    # Écrire les données modifiées dans le fichier JSON
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)




remove_fields_from_splits("sorted_London_2024.json")
```

## Splitter la nationalité 

```python
import json
import re

def split_name_details(file_path):
    # Lire le fichier JSON
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Séparer les champs Name, Surname et Nationality
    for runner_id, runner_data in data.items():
        if "Participant Details" in runner_data:
            name_ctz = runner_data["Participant Details"].get("Name (CTZ)", "")
            match = re.match(r"([^,]+),\s*([^(]+)\s*\((\w+)\)", name_ctz)
            if match:
                name = match.group(1).strip()
                surname = match.group(2).strip()
                nationality = match.group(3).strip()
            else:
                name = ""
                surname = name_ctz.strip()
                nationality = "False"
            
            runner_data["Participant Details"]["Name"] = name
            runner_data["Participant Details"]["Surname"] = surname
            runner_data["Participant Details"]["Nationality"] = nationality
            del runner_data["Participant Details"]["Name (CTZ)"]
    
    # Écrire les données modifiées dans le fichier JSON
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

# Utilisation de la fonction
split_name_details('sorted_London_2024.json')
```

# changer le nom de la clé temps_total

```python
import json

def rename_finish_time_key(file_path):
    # Lire le fichier JSON
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Renommer la clé "Finish Time" en "Time Total (net)" dans "Totals"
    for runner_id, runner_data in data.items():
        if "Totals" in runner_data:
            totals = runner_data["Totals"]
            if "Place (Overall)" in totals:
                totals["Place (Total)"] = totals.pop("Place (Overall)")
    
    # Écrire les données modifiées dans le fichier JSON
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

#Utilisation de la fonction
rename_finish_time_key('sorted_London_2024.json')
```

# changer les noms des splits

```python
import json

def replace_split_names(file_path):
    # Lire le fichier JSON
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Remplacer les valeurs de "Split"
    for runner_id, runner_data in data.items():
        if "Splits" in runner_data:
            for split in runner_data["Splits"]:
                if split["Split"] == "Half":
                    split["Split"] = "HALF"
                elif split["Split"] == "Finish":
                    split["Split"] = "Finish Net"

    # Remplacer les valeurs de "Split"
    for runner_id, runner_data in data.items():
        if "Participant Details" in runner_data:
            if "Gender" in runner_data["Participant Details"]:
                runner_data["Participant Details"]["Sex"] = runner_data["Participant Details"]["Gender"]
            

    
    # Écrire les données modifiées dans le fichier JSON
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

# Utilisation de la fonction
replace_split_names('sorted_Chicago_2023.json')





```
